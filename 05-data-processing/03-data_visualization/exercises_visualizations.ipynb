## 1. Counting birds

### <u> Counting birds part 1 </u>

Every year there is a day where people are encouraged to count the birds in their garden. Joseph tries to participate every year. Below you see the data of Joseph's observations. Let's make a nice plot of the data, also include the total amount of birds spotted.

```python
# Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'others': [104, 98, 88, 92, 101]
}
```

### Inspecting the data

Load and inspect the data.

```python
# Load and inspect
df = pd.DataFrame(data)
df
```

### Plotting the pigeon count

Plot the number of counted pigeons over the year, following these specifications:

* Line graph
* Size of the figure: (10,6)
* Full blue line, data point marked with a circle, linewidth 2, marker size 8
* Provide the necessary labels in order to interpret the graph and show a legend
* Provide a meaningfull title
* Show a grid

```python
#Plotting the Pigeon count

plt.figure(figsize = (10, 6))

plt.plot(df["year"], df["Pigeons"], marker = "o", color = "blue", linestyle = "-", linewidth = 2, markersize = 8, label = "Pigeons")
plt.title("Pigeons over the year")
plt.xlabel("years")
plt.ylabel("pigeons")
plt.legend()
plt.grid(True)

plt.show()
```

### Plotting the sparrow count

Plot the number of sparrows Joseph counted over the years, adhere to the specifications listed below:

* Line graph
* linestyle: dotted
* color of the graph: Green
* marker: a triangle pointing to the right
* the edge of the markers should be red with width 1
* Corresponding labels and titles

Ensure the plot of the Pigeon count is incorporated as well.


```python
# Plotting the sparrow count

plt.figure(figsize=(10, 6))

plt.plot(
    df["year"],
    df["Sparrows"],
    marker = ">",
    linestyle = ":",
    color = "green",
    markeredgecolor = "red",
    markeredgewidth = 1,
    linewidth = 2,
    markersize=8,
    label="Sparrows"
)

plt.xlabel("Years")
plt.ylabel("Sparrows")
plt.title("Sparrow count over the years")
plt.legend()
plt.grid(True)
plt.show()
```

### Plotting the other birds

Add a graph line for all other birds counted. You can use a styling of your choice.

```python
# Plotting the other birds count

plt.figure(figsize=(10,6))

plt.plot(
    df['year'],
    df['others'],
    marker="o",
    linestyle="-.",
    color="pink",
    linewidth=2,
    markersize=8,
    label="other birds"
)

plt.xlabel("years")
plt.ylabel("other birds")
plt.title("Other birds over the years")
plt.legend()
plt.grid(True)

plt.show()
```

### <u> Counting birds part 2 </u>

### Plotting the total of birds counted

Add a column to the dataframe containing the total amount of birds counted that year.


```python
# Total of birds
df["total"] = df["Pigeons"] + df["Sparrows"] + df["others"]

df
```



Add a **barplot** for the total of birds counted that day on top of the already created plots of the bird species. Some specifications:
* Bar width = 0.4
* Bar color = purple
* Use corresponding labels 

Use corresponding labels 

```python
# Write your code here

plt.figure(figsize=(10, 6))

plt.bar(
    df["year"],
    df["total"],
    width=0.4,
    color="purple"
)

plt.xlabel("Year")
plt.ylabel("Total birds counted")
plt.title("Total birds counted per year")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### <u> Counting birds part 3 </u>

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
# Your code
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# --------- Left Plot: Stacked Bar Chart ---------
axes[0].bar(df["year"], df["Pigeons"], color="lightblue", label="Pigeons")
axes[0].bar(
    df["year"],
    df["Sparrows"],
    bottom=df["Pigeons"],
    color="lightgreen",
    label="Sparrows",
)
axes[0].bar(
    df["year"],
    df["others"],
    bottom=df["Pigeons"] + df["Sparrows"],
    color="lightcoral",
    label="Others",
)

axes[0].set_title("Birds Counted Per Year (Stacked)")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Number of Birds")
axes[0].legend()
axes[0].grid(axis="y", linestyle="--", alpha=0.5)

# --------- Right Plot: Total Bar Chart ---------
axes[1].bar(df["year"], df["total"], color="purple", width=0.4)

axes[1].set_title("Total Birds Counted Per Year")
axes[1].set_xlabel("Year")
axes[1].grid(axis="y", linestyle="--", alpha=0.5)

# --------- General Styling ---------
fig.suptitle(
    "Overview of Bird Counts from 2020 to 2024", fontsize=16, fontweight="bold"
)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
# Your code
years = df["year"]
x = np.arange(len(years))  # the label locations
bar_width = 0.18

plt.figure(figsize=(12, 6))

# Grouped bars
plt.bar(
    x - bar_width * 1.5,
    df["Pigeons"],
    width=bar_width,
    label="Pigeons",
    color="#4C72B0",
)
plt.bar(
    x - bar_width * 0.5,
    df["Sparrows"],
    width=bar_width,
    label="Sparrows",
    color="#55A868",
)
plt.bar(
    x + bar_width * 0.5, df["others"], width=bar_width, label="Others", color="#C44E52"
)
plt.bar(
    x + bar_width * 1.5, df["total"], width=bar_width, label="Total", color="#8172B2"
)

# Labels and formatting
plt.xlabel("Year", fontsize=12)
plt.ylabel("Bird Count", fontsize=12)
plt.title("Bird Counts Grouped by Category per Year", fontsize=14, weight="bold")
plt.xticks(ticks=x, labels=years)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
